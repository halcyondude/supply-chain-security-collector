# Overview of Aggregated (agg_) Tables
This notebook displays all major `agg_` tables from the DuckDB database, including SBOMs found in both release artifacts and security insights files.

**SBOMs are detected in two main ways:**
- **Release Artifacts:** SBOMs attached to GitHub releases.
- **Security Insights Files:** SBOMs found via security scanning or insights files.

Each section below queries and displays a different `agg_` table.

```python
import duckdb
import pandas as pd

# Path to DuckDB database (update if needed)
DB_PATH = '../output/cncf-full-landscape/current/database.db'

# Connect to DuckDB (read-only)
conn = duckdb.connect(DB_PATH, read_only=True)
print(f"Connected to DuckDB: {DB_PATH}")
```

```python
# Helper function to display all rows of a table as a DataFrame
def display_table(conn, table_name):
    try:
        df = conn.execute(f'SELECT * FROM "{table_name}"').df()
        print(f"Table: {table_name} ({len(df)} rows):")
        display(df)
    except Exception as e:
        print(f"{table_name}: error displaying table - {e}")
```

```python
all_tables = [t[0] if isinstance(t, (list, tuple)) else t for t in conn.execute("SHOW TABLES").fetchall()]
all_tables_df = pd.DataFrame(all_tables, columns=['table_name'])
display(all_tables_df)
```

```python
for table_name in all_tables:
    display_table(conn, table_name)
```
